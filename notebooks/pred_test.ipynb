# Build train, test, val sequences

```python
df1 = pd.read_pickle("../full_tracks_1.pkl") 
df2 = pd.read_pickle("../full_tracks_2.pkl") 
df3 = pd.read_pickle("../full_tracks_3.pkl")
df4 = pd.read_pickle("../full_tracks_4.pkl") 
df5 = pd.read_pickle("../full_tracks_5.pkl") 
df6 = pd.read_pickle("../full_tracks_6.pkl")  
```

```python
track_df = pd.concat([df1, df2], axis=0)
track_df = pd.concat([track_df, df3], axis=0)
track_df = pd.concat([track_df, df4], axis=0)
track_df = pd.concat([track_df, df5], axis=0)
track_df = pd.concat([track_df, df6], axis=0)
```
